```python
#CPI means increase of % compared to 1982 - 100 . Ex: %250 = 150% difference
#Average cost of living in 1982 ~670/month rough estimate

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
infileCoL = "CoL.csv"
CoL = pd.read_csv(infileCoL, index_col=0)[['YrCol','Yearly']]

checkYear = 2025 # Change this value to check estimated year
```

```python
plt.plot(CoL)
plt.xlabel("Year")
plt.ylabel("Yearly Cost of Living")
```

```python
yYearly = np.array((CoL.Yearly))
yCoL = np.array((CoL.YrCol))
x = np.array(list(range(1938,2021))).reshape((-1,1))
model = LinearRegression().fit(x,yYearly)
modelCoL = LinearRegression().fit(x,yCoL)
```

```python
# print(yYearly)
# print(yCoL)
```

```python
satYearly = model.score(x,yYearly)
satCoL = modelCoL.score(x,yCoL)
# print('coefficient of determination: ', satYearly)
# print('coefficient of determination: ', satCoL)
```

```python
# print('intercept', model.intercept_)
# print('slope', model.coef_)
# print('intercept', modelCoL.intercept_)
# print('slope', modelCoL.coef_)
```

```python
y_pred = model.predict(x)
y_predCoL = modelCoL.predict(x)
# print('predicted response: ', y_pred, sep= '\n')
# print('predicted response: ', y_predCoL, sep= '\n')
```

```python
x_new = np.arange(1938, checkYear).reshape((-1,1))
y_new = model.predict(x_new)
y_newYr = modelCoL.predict(x_new)
# print(x_new)
```

```python
plt.plot(model.predict(x_new))
plt.plot(modelCoL.predict(x_new))
plt.xlabel("Years since 1938")
plt.ylabel("Annual Cost of Living")

print(str(checkYear) + " estimated salary based on min wage: " + str(round(y_new[len(y_new)-1], 2)))
print(str(checkYear) + " estimated cost of living: " + str(round(y_newYr[len(y_new)-1], 2)))
```
